# Telco Customer Churn — Feature Engineering

## Goal
Prepare the data for modeling:
- Drop unnecessary columns
- Encode categorical variables
- Handle class imbalance
- Save processed data

## 1. Imports & Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('../data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv')
print(df.shape)
df.head()
```

## 2. Drop Unnecessary Columns

```python
# customerID is just an identifier — no predictive value
df = df.drop(columns=['customerID'])
print(f"Shape after drop: {df.shape}")
```

## 3. Encode Target Variable

```python
# Churn: Yes=1, No=0
df['Churn'] = (df['Churn'] == 'Yes').astype(int)
print(df['Churn'].value_counts())
```

## 4. Encode Categorical Variables

```python
# Fix TotalCharges — stored as object
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'] = df['TotalCharges'].fillna(0)
```

```python
# Binary: Yes=1, No=0
binary_cols = ['Partner', 'Dependents', 'PhoneService', 'PaperlessBilling']
for col in binary_cols:
    df[col] = (df[col] == 'Yes').astype(int)

# gender: Male=1, Female=0
df['gender'] = (df['gender'] == 'Male').astype(int)

# Churn: Yes=1, No=0
df['Churn'] = (df['Churn'] == 'Yes').astype(int)

print("Binary encoding done")
```

```python
# These columns have 3 values including 'No internet service' / 'No phone service'
# Treat 'No internet service' and 'No phone service' same as 'No' → 0
three_val_cols = ['MultipleLines', 'OnlineSecurity', 'OnlineBackup',
                  'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

for col in three_val_cols:
    df[col] = (df[col] == 'Yes').astype(int)

print("Three-value encoding done ")
```

```python
# One-Hot Encoding for multi-class columns 
df = pd.get_dummies(df, columns=['InternetService', 'Contract', 'PaymentMethod'], drop_first=False)

print(f"Shape after encoding: {df.shape}")
df.head()
```

```python
# Verify no object columns remain
remaining = df.select_dtypes(include='object').columns.tolist()
print(f"Remaining object columns: {remaining}")
```

## 5. Save Processed Data

```python
df.to_csv('../data/processed/telco_processed.csv', index=False)
print("Processed data saved ")
print(f"Final shape: {df.shape}")
df.head()
```

## 6. Conclusions

- Dropped `customerID` — no predictive value
- Fixed `TotalCharges` — converted from string to numeric
- Binary encoded: `gender`, `Partner`, `Dependents`, `PhoneService`, `PaperlessBilling`
- Three-value encoded: `MultipleLines`, `OnlineSecurity`, `OnlineBackup`, `DeviceProtection`, `TechSupport`, `StreamingTV`, `StreamingMovies`
- One-Hot encoded: `InternetService`, `Contract`, `PaymentMethod`
- Final dataset: **7,043 rows × 27 columns** — ready for modeling
